### Create a dummy dataframe  

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("DuplicateCheck").getOrCreate()

data = [
    (1, "A", 100),
    (1, "A", 100),
    (2, "B", 200),
    (3, "C", 300),
    (2, "B", 200),
    (4, "D", 400)
]

columns = ["id", "category", "amount"]

df = spark.createDataFrame(data, columns)
df.show()
```

### identify duplicates using composite key

```python
duplicate_df = (
    df.groupBy("id", "category")
      .count()
      .filter(col("count") > 1)
)

duplicate_df.show()
```

### Extract Full Duplicate Records (Optional)

```python
duplicates_full = df.join(
    duplicate_df.select("id", "category"),
    on=["id", "category"],
    how="inner"
)

duplicates_full.show()
```

### Unit Test (PySpark + pytest style)

```python
def test_duplicates(spark_session):
    data = [
        (1, "A", 100),
        (1, "A", 100),
        (2, "B", 200)
    ]
    columns = ["id", "category", "amount"]

    df = spark_session.createDataFrame(data, columns)

    result = (
        df.groupBy("id", "category")
          .count()
          .filter(col("count") > 1)
    )

    assert result.count() == 1, f"Test Failed!"

    print("✅ Test Passed!")
```

```python
test_duplicates(spark)
```
