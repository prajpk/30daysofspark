```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, dense_rank
from pyspark.sql.window import Window

spark = SparkSession.builder.appName("TopEmployees").getOrCreate()

# Dummy Data
data = [
    (1, "Alice", "HR", 50000),
    (2, "Bob", "HR", 60000),
    (3, "Charlie", "HR", 55000),
    (4, "David", "IT", 70000),
    (5, "Eve", "IT", 80000),
    (6, "Frank", "IT", 75000),
    (7, "Grace", "Finance", 65000),
    (8, "Heidi", "Finance", 72000),
    (9, "Ivan", "Finance", 71000)
]

columns = ["emp_id", "emp_name", "department", "salary"]

df = spark.createDataFrame(data, columns)

# Window Spec
window_spec = Window.partitionBy("department").orderBy(col("salary").desc())

# Apply Ranking
ranked_df = df.withColumn("rank", dense_rank().over(window_spec))

# Filter Top 3
top_3_df = ranked_df.filter(col("rank") <= 3)

top_3_df.show()
```
