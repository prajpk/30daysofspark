```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, dense_rank
from pyspark.sql.window import Window

spark = SparkSession.builder.appName("SecondHighestDeptSalary").getOrCreate()

# Sample data
data = [
    (1, "Alice", "HR", 5000),
    (2, "Bob", "HR", 7000),
    (3, "Charlie", "HR", 7000),
    (4, "David", "IT", 9000),
    (5, "Eve", "IT", 8000),
    (6, "Frank", "IT", 8000),
    (7, "Grace", "Finance", 6000)
]

columns = ["id", "name", "department", "salary"]

df = spark.createDataFrame(data, columns)

# Window spec partitioned by department
window_spec = Window.partitionBy("department").orderBy(col("salary").desc())

# Apply dense rank
ranked_df = df.withColumn("rank", dense_rank().over(window_spec))

# Filter second highest salary per department
second_highest_df = ranked_df.filter(col("rank") == 2)

second_highest_df.show()
```
